```python
for i in range(1,len(scores)):
    scores[i] = eval(scores[i])
```

```python
scores.drop(index=0, inplace=True)
```

```python
for i in range(1, len(scores)-1):
    if len(scores[i]) == 1:
        new_scores = new_scores.append(scores[i], ignore_index=True)
    elif scores[i] == 'ScoringDetails':
        break
    else:
        for score in scores[i]:
            new_scores = new_scores.append(score, ignore_index=True)
```
